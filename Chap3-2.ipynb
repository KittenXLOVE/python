## List[串列] Tuple(元組) Dictionary{字典} Sets{集合}

```python
ListA=["A","B","C"]
TupleA=("A","B","C")
TupleB=tuple(ListA)#轉換成Tuple
print(ListA)
print(TupleA)
print(TupleB)
```

```python
ListA=["A","B","C"]#List
dictA={0:"A",1:"B"}#Dictionary
dictB={"name":"Apple","color":"Blue"}#Dictionary
print(ListA)
print(dictA)
print(dictB)
print(dictB["name"])#Dictionary取值
```

```python
dictA={"name":"Apple","color":"Blue"}
print(dictA.keys())
print(dictA.values())
print(dictA.items())
print("name" in dictA)#判斷"name"是否存在Dictionary中
```

```python
import random
dice=random.randint(1,6)#亂數產生1~6
print("dice=",dice)
if dice==1:
  print("壹點")
elif dice==2:
  print("貳點")
elif dice==3:
  print("參點")
elif dice==4:
  print("肆點")
elif dice==5:
  print("伍點")
elif dice==6:
  print("陸點")
```

```python
import random
dice=random.randint(1,6)
print("dice=",dice)
dice_cht=["壹","貳","參","肆","伍","陸"]
print(dice_cht[dice]+"點")#使用陣列簡化程式碼
```
