# Chap 7 錯誤處理

```python
x=5
y=50
if x<y:
  print(z)#無宣告的變數，發生錯誤
```

```python
x=5
y=50
if x<y:
  try:#使用try偵測錯誤
    print(z)
  except:#發生錯誤時會進except
    print("錯誤！")
```

```python
x=5
y=50
if x<y:
  try:
    print(z)
  except NameError as e:#except可個別指定錯誤原因
    print("錯誤！原因：",e)
  except ZeroDivisionError:
    print("嘗試除以零")
```
