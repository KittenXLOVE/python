## 為什麼需要numpy?
* numpy由c撰寫，速度快python 10~1000倍以上
* numpy儲存結果較正確，python數字儲存不正確
* numpy節省記憶體儲存空間
* numpy的向量(矩陣計算)速度快
* numpy的取值方便

## Numpy的乘法

```python
import numpy as np
source = [1,2,3,4,5]
array=np.array(source)#轉換成numpy陣列
print(source*4)
print(array*4)#會對每筆資料做乘法，而不是乘以幾次的資料
```

## ndarray.shape 陣列維度的形狀
## ndarray.dtype 陣列的資料型態
## ndarray.ndim 陣列的維度

```python
import numpy as np
array=np.array([1,2,3,4,5])
print(type(array))#取得變數型態
print(array.shape)
print(array.dtype)
print(array.ndim)
```

## numpy.zeros 產生資料為0的陣列
## numpy.empty 產生資料為空的陣列
## numpy.arange 產生指定範圍、間距的等差陣列

```python
import numpy as np
a = np.zeros(6)#產生6筆資料為零的陣列
b = np.empty((2,3,4))#產生資料為空的2*3*4之3維陣列
c = np.arange(10)#產生0~10的陣列，另可設定起始值及間隔值

print(a)
print(b)
print(c)
```

## 2維陣列取值

```python
import numpy as np
a = np.array((np.arange(0,3),np.arange(7,10)))

print(a)
print(a[0,:])#取所有在1維中index為0的資料
print(a[:,0])#取所有在2維中index為0的資料
print(a[1,1])#取1維index為1且2維index為1的資料
```

## ndarray.reshape 重新定義陣列維度形狀(須符合舊有資料的數量)

```python
import numpy as np
a = np.arange(10)
print(a)
print(a.shape)
a = a.reshape(2,5)
print(a)#資料會隨著陣列大小重新編排
print(np.arange(100).reshape(10,10))
```

## numpy.nan 空值

```python
import numpy as np
a = np.array([10,54,17,56,np.nan])
print(a)
```
